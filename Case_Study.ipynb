```python
#import the necessary libraries for the case study
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load dataset
data=pd.read_csv("/content/train_loan.csv")
data
```

```python
data.info()
```

```python
data.describe()
```

```python
#Check for duplicate values
data.duplicated().sum()
```

```python
#check for missing values
data.isnull().sum()
```

```python
#check the percentage of null values
round(data.isnull().mean()*100,2)
```

```python
#separating numerical and categorical values
num_cols = data.select_dtypes(include=['int64', 'float64']).columns
cat_cols = data.select_dtypes(include=['object']).columns
```

```python
#filling numerical columns with median
for col in num_cols:
    data[col].fillna(data[col].median(), inplace=True)
```

```python
#filling categorical columns with mode
for col in cat_cols:
    data[col].fillna(data[col].mode()[0], inplace=True)

```

```python
#checking null values after imputation
data.isnull().sum()
```

```python
#checking outliers
Q1 = data[num_cols].quantile(0.25)
Q3 = data[num_cols].quantile(0.75)
IQR = Q3 - Q1

outliers = ((data[num_cols] < (Q1 - 1.5 * IQR)) |
            (data[num_cols] > (Q3 + 1.5 * IQR))).sum()

outliers
```

```python
#capping
for col in num_cols:
    lower = Q1[col] - 1.5 * IQR[col]
    upper = Q3[col] + 1.5 * IQR[col]
    data[col] = np.clip(data[col], lower, upper)
```

```python
# start encoding for categorical values
cat_cols = data.select_dtypes(include=['object']).columns
cat_cols
```

```python
#one hot encoding
data = pd.get_dummies(data, columns=cat_cols, drop_first=True, dtype=int)

```

```python
data.head()
```

```python
#scaling
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
data[num_cols] = scaler.fit_transform(data[num_cols])
```
